# VISUALIZACIONES

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el archivo CSV con los datos
file_path = "data/2024_Accidentalidad.csv"
data = pd.read_csv(file_path, encoding='latin1', delimiter=';', on_bad_lines='skip', engine='python')

# Convertir las columnas a formatos adecuados
data['fecha'] = pd.to_datetime(data['fecha'], format='%d/%m/%Y', errors='coerce')
data['hora'] = pd.to_datetime(data['hora'], format='%H:%M:%S', errors='coerce').dt.hour

# Crear una columna para d√≠a de la semana
data['dia_semana'] = data['fecha'].dt.day_name()

# Categorizar accidentes con alcohol o droga implicados
data['accidente_alcohol_droga'] = data.apply(lambda row: 'Con Alcohol/Droga' if row['positiva_alcohol'] == 'S' or row['positiva_droga'] == 'S' else 'Sin Alcohol/Droga', axis=1)

# Configurar estilo y paleta de colores para las visualizaciones
sns.set_palette("viridis")
plt.style.use('ggplot')
```

```python
# Gr√°fico: N√∫mero de Accidentes por Distrito
plt.figure(figsize=(12, 8))
sns.countplot(data=data, y='distrito', order=data['distrito'].value_counts().index, palette='viridis')
plt.title('N√∫mero de Accidentes por Distrito', fontsize=16, weight='bold')
plt.xlabel('Cantidad de Accidentes', fontsize=14)
plt.ylabel('Distrito', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Gr√°fico 1: N√∫mero de Accidentes por Tipo de Veh√≠culo
plt.figure(figsize=(12, 8))
sns.countplot(data=data, x='tipo_vehiculo', palette='viridis', order=data['tipo_vehiculo'].value_counts().index)
plt.title('N√∫mero de Accidentes por Tipo de Veh√≠culo', fontsize=16, weight='bold')
plt.xlabel('Tipo de Veh√≠culo', fontsize=14)
plt.ylabel('Cantidad de Accidentes', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Gr√°fico 2: Accidentes con Alcohol/Drogas vs Sin Alcohol/Drogas
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='accidente_alcohol_droga', palette='viridis')
plt.title('N√∫mero de Accidentes con Alcohol/Drogas y Sin Ellos', fontsize=16, weight='bold')
plt.xlabel('Categor√≠a de Accidentes', fontsize=14)
plt.ylabel('Cantidad de Accidentes', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Gr√°fico 3: Horas M√°s Frecuentes de Accidentes
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='hora', bins=24, kde=True, color='teal')
plt.title('Horas M√°s Frecuentes de Accidentes', fontsize=16, weight='bold')
plt.xlabel('Hora del D√≠a', fontsize=14)
plt.ylabel('Cantidad de Accidentes', fontsize=14)
plt.xticks(ticks=range(0, 24), labels=[f"{i}h" for i in range(0, 24)], fontsize=12)  # Ajustar etiquetas
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Gr√°fico 4: Frecuencia de Accidentes por D√≠a de la Semana
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='dia_semana', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], palette='viridis')
plt.title('Frecuencia de Accidentes por D√≠a de la Semana', fontsize=16, weight='bold')
plt.xlabel('D√≠a de la Semana', fontsize=14)
plt.ylabel('Cantidad de Accidentes', fontsize=14)
plt.xticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Gr√°fico 5: Edades de Implicados en Accidentes
plt.figure(figsize=(12, 6))
sns.countplot(data=data, y='rango_edad', order=data['rango_edad'].value_counts().index, palette='viridis')
plt.title('Edades de Implicados en Accidentes', fontsize=16, weight='bold')
plt.xlabel('Cantidad de Implicados', fontsize=14)
plt.ylabel('Rango de Edad', fontsize=14)
plt.xticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Gr√°fico: N√∫mero de Accidentes por Condiciones Meteorol√≥gicas
plt.figure(figsize=(12, 8))
sns.countplot(data=data, y='estado_meteorol√≥gico', order=data['estado_meteorol√≥gico'].value_counts().index, palette='viridis')
plt.title('N√∫mero de Accidentes por Condiciones Meteorol√≥gicas', fontsize=16, weight='bold')
plt.xlabel('Cantidad de Accidentes', fontsize=14)
plt.ylabel('Condiciones Meteorol√≥gicas', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Gr√°fico 1: N√∫mero de Accidentes por Sexo de la Persona Involucrada
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='sexo', palette='viridis')
plt.title('N√∫mero de Accidentes por Sexo de la Persona Involucrada', fontsize=16, weight='bold')
plt.xlabel('Sexo', fontsize=14)
plt.ylabel('Cantidad de Accidentes', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Gr√°fico 2: N√∫mero de Accidentes por Rol de la Persona Involucrada (tipo_persona)
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='tipo_persona', palette='viridis')
plt.title('N√∫mero de Accidentes por Rol de la Persona Involucrada', fontsize=16, weight='bold')
plt.xlabel('Rol de la Persona Involucrada', fontsize=14)
plt.ylabel('Cantidad de Accidentes', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Gr√°fico 3: N√∫mero de Accidentes por Tipo de Accidente
plt.figure(figsize=(12, 8))
sns.countplot(data=data, y='tipo_accidente', order=data['tipo_accidente'].value_counts().index, palette='viridis')
plt.title('N√∫mero de Accidentes por Tipo', fontsize=16, weight='bold')
plt.xlabel('Cantidad de Accidentes', fontsize=14)
plt.ylabel('Tipo de Accidente', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
